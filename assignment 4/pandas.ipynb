```python
import pandas as pd
from matplotlib import pyplot as plt 
```

```python
gross_product = pd.read_csv('gross.csv')
gross_product
```

```python
plt.plot(gross_product.label, gross_product.GDP_Per_Capita)
plt.title('Gross domestic product per capita in the World')
plt.xlabel('year')
plt.ylabel('GDP per capita')
plt.show()
```

```python
plt.scatter(gross_product.label, gross_product.GDP_Per_Capita)
plt.title('Gross domestic product per capita in the World')
plt.xlabel('year')
plt.ylabel('GDP per capita')
plt.show()
```

```python
urban_rural = pd.read_csv('urban_rural.csv')
urban_rural
```

```python
plt.plot(urban_rural.label, urban_rural.Urban_Population / 10 ** 9)
plt.plot(urban_rural.label, urban_rural.Rural_Population / 10 ** 9)
plt.title('Urban and rural population in the World')
plt.legend(['Urban', 'Rural'])
plt.xlabel('year')
plt.ylabel('population (billion)')
plt.show()
```

```python
urban = pd.read_csv('urban.csv')
df = pd.DataFrame(urban)
urban
```

```python
x = list(df.iloc[:, 0])
y = list(df.iloc[:, 1])

plt.bar(x, y)
plt.title('Ranking by Population (Urban)')
plt.ylabel('population (100 million)')
plt.show()
```

```python
rural = pd.read_csv('rural.csv')
rural
```

```python
country = rural['country']
value = rural['value']
colors = ["#1f77b4", "#ff7f0e", "#2ca02c", "#d62728", "#8c564b"]
plt.title('Ranking by Population (Rural)')
plt.pie(value, labels=country, colors=colors,
autopct='%1.1f%%')
plt.show()
```
